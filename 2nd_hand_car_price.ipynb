```python
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set()
from sklearn.linear_model import LinearRegression
```

```python
raw_data=pd.read_csv('car_sales_2.csv')
raw_data.describe()
raw_data.head()
```

```python
data=raw_data.drop(['Model'],axis=1)
data.describe(include="all")
```

```python
data.isnull().sum()
```

```python
data=data.dropna(axis=0)
data.isnull().sum()
```

```python
data.describe(include='all')
```

```python
sns.displot(data['Price'])
```

```python
q=data['Price'].quantile(0.99)
data=data[data['Price']<q]
data.describe(include='all')
```

```python
sns.distplot(data['Price'])
```

```python
sns.distplot(data['Mileage'])
```

```python
q=data['Mileage'].quantile(0.99)
data=data[data['Mileage']<q]
```

```python
data.describe(include='all')
```

```python
data=data[data['EngineV']<6.5]
data.describe(include='all')
```

```python
data['EngineV']
```

```python
data['EngineV'].max()
```

```python
q=data['Year'].quantile(0.01)
data=data[data['Year']>q]
```

```python
sns.distplot(data['Year'])
```

```python
data.describe(include='all')
```

```python
log_price=np.log(data['Price'])
data['log_price']=log_price
data
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables=data[['Mileage','Year','EngineV']]
vif=pd.DataFrame()
vif['vif']=[variance_inflation_factor(variables.values,i)for i in range(variables.shape[1])]
```

```python
vif['features']=variables.columns
```

```python
vif
```

```python
data=data.drop(['Year'],axis=1)
```

```python
data=pd.get_dummies(data,drop_first=True)
```

```python
data
```

```python
data.columns.values
```

```python
data=data.drop(['Price'],axis=1)
```

```python
data
```

```python
data.columns.values
```

```python
cols=['log_price','Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_proprocess=data[cols]
```

```python
data_proprocess.head()
```

```python
data_preprocess=data_proprocess
data_preprocess.head()
```

```python
target=data_preprocess['log_price']
inputs=data_preprocess.drop(['log_price'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(inputs)
inputs_scaled=scaler.transform(inputs)
```

```python
inputs_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs_scaled,target,test_size=0.2, random_state=34)
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
y_hat=reg.predict(x_train)
plt.scatter(y_hat,y_train)
plt.xlabel("y_hat")
plt.ylabel("y train")
plt.show()
```

```python
sns.displot(y_train-y_hat)
```

```python
reg.score(x_train,y_train)
```

```python
reg_summary=pd.DataFrame(inputs.columns.values,columns=['features'])
reg_summary['weights']=reg.coef_
```

```python
reg_summary
```

```python
y_hat_test=reg.predict(x_test)
plt.scatter(y_hat_test,y_test)
plt.xlabel("y hat test")
plt.ylabel("y test")
plt.show()
```

```python
df_pf=pd.DataFrame(np.exp(y_hat_test),columns=['predictions'])
df_pf.head()
```

```python
y_test=y_test.reset_index(drop=True)
```

```python
df_pf['target']=np.exp(y_test)
df_pf
```

```python
df_pf['residual']=df_pf['target']-df_pf['predictions']
```

```python
df_pf
```

```python
df_pf['residual'].max()
```

```python
df_pf['residual'].min()
```

```python
df_pf['percentage%']=np.absolute((df_pf['residual']/df_pf['target'])*100)
df_pf
```

```python
df_pf.describe()
```
